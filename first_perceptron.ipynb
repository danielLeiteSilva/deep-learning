```python
import numpy as np
```

```python
#Tabala de entrada
entradas = np.array([[0,0], [0,1], [1,0], [1,1]])
```

```python
#Tabela de classe para comparar resultados com o esperado no começo
#Resultados esperados do modelo
saidas = np.array([0,1,1,0]) #Operador OU  
```

```python
saidas
```

```python
#Definir pesos iniciais
pesos = np.array([0.0, 0.0])
```

```python
#Taxa de aprendizagem
txAprendizagem = 0.1
```

```python
#Funcao Soma
def somaFunction(e):
  return e.dot(pesos)
```

```python
#Definir funcao ativação
def stepFunction(numero):
  if(numero >= 1):
    return 1
  return 0
```

```python
#Treinar o modelo
def treinarModelo():
  erroTotal = 1

  #Equanto o erro for diferente
  while(erroTotal != 0):
    erroTotal = 0
    for i in range(len(saidas)):
      #Funcao para aplicar os pesos nas entradas
      soma = somaFunction(np.asarray(entradas[i]))

      #Chamada da funcao de ativação
      saidaCalculada = stepFunction(soma)

      #Calcula o erro
      erro = abs(saidas[i] - saidaCalculada)

      #Adiciono resultado no erro total
      erroTotal += erro

      #Realizar os ajustes nos pesos
      for j in range(len(pesos)):
        pesos[j] = pesos[j] + (txAprendizagem * entradas[i][j] * erro)
        print('Peso atualizado: ' + str(pesos[j]))

    print('Total Erros: ' + str(erroTotal))
```

```python
treinarModelo()
print("Rede neural treinada")
```
